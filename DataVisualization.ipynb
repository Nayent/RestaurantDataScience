```
from scipy.stats import ranksums

valor_conta_domingo = gorjeta.query("Dia == 'Domingo'").Valor_Total
valor_conta_sabado = gorjeta.query("Dia == 'Sabado'").Valor_Total
ranksums(valor_conta_domingo, valor_conta_sabado)
```

#Com o pvalue < 0.05 temos que são distintos, porém por ser maior eles são parecidos ou iguais
